**Descrição de colunas:**

* `survival`: variável binária que representa se o indivíduo sobreviveu (1) ou não (0);
* `pclass`: em qual classe ele estava alojado;
* `name`: nome do passageiro;
* `sex`: gênero do passageiro;
* `age`: idade do passageiro;
* `sibsp`: número de esposos e/ou irmãos à bordo;
* `parch`: número de pais ou filhos à bordo;
* `ticket`: número da passagem;
* `fare`: preço pago pela passagem;
* `cabin`: número da cabine em que estava alojado;
* `embarked`: cidade em que embarcou.

# **Criação do Dataframe**

**Etapa de coleta de dados**
Vamos analisar a fonte de dados em csv, assim precisamos criar um dataframe ao ler o arquivo dado : titanic.csv

```python
dataframe = pd.read_csv("titanic.csv")
dataframe
```

## **Tratando os dados**

```python
#Retorna no número de linhas e colunas
dataframe.shape
```

```python
#Retorna no nome de cada coluna do dataframe
dataframe.columns
dataframe.rename(columns={'Survived': 'Sobreviveu'})
```

```python
#Alteração de nome da coluna Survived e Name
dataframe_renomeado = dataframe.rename(columns={'Survived': 'Sobreviveu', 'Name': 'Nome'})
dataframe_renomeado
```

```python
dataframe_renomeado.columns
```

```python
#Soma os valores nulos e exibe-os em forma de números. Caso execute apenas o isnull(), retornará um dataframe mostrando se aquela
#coluna é true ou false.
dataframe.isnull().sum()
```

```python
#Soma e mostra quantia de passageiros que sobreviveram 
dataframe['Survived'].sum()
```

```python
#Dataframe cópia, para realizar deletes, experimentos, explorar...
dataframe_copia = dataframe.copy()
```

```python
dataframe_copia.rename(columns= {"Parch": "Parceiro"})
```

```python
#Excluindo uma coluna
dataframe_copia.drop(['SibSp', 'Parch'], axis='columns', inplace=True)
```

```python
dataframe_copia
```

```python
#Excluindo valores duplicados
dataframe_copia.drop_duplicates(inplace=True)
dataframe_copia.shape
```

## **Tabela até o momento**

- Importamos o arquivo CSV com os dados 
- Visualizamos a tabela 
- Verificamos o tamanho dela
- Verificamos quantia de valores nulos
- Excluímos duas colunas das quais não faziam sentido para o objetivo da análise

Abaixo, tabela até o momento: 

```python
dataframe_copia.head(10)
```

```python
#Contagem de quantia de homens x mulheres
contagem_genero = dataframe_copia['Sex'].value_counts()
contagem_genero
```

```python
#Gerando gráficos
contagem_genero.plot(kind='bar', color='pink')
plt.xlabel('Genero')
plt.ylabel('Quantidade')
plt.title('Quantidade de passageiros por Gênero')
```

É possível observar que haviam mais homens do que mulheres a bordo do Titanic. 

```python
#Média de idade por sexos
contagem_idade = dataframe_copia.groupby('Sex')['Age'].mean()
contagem_idade
```

```python
contagem_idade.plot(kind='bar', color='pink')
plt.xlabel('Idade')
plt.ylabel('Quantidade')
plt.title('Média de passageiros por Gênero')
```

```python
#Criando uma nova coluna, para adicionar as faixas etárias dos passageiros abordo. 
bins = [0, 11, 19, 60, 100]
labels = ['Criança', 'Adolescente', 'Adulto', 'Idoso']

dataframe_copia['FaixaEtaria'] = pd.cut(dataframe_copia['Age'], bins=bins, labels=labels, right=False, include_lowest=True)
dataframe_copia   
```

```python
# Quantidade de Passageiros por Faixa Etária
contagem_faixa_etaria = dataframe_copia['FaixaEtaria'].value_counts()
print(contagem_faixa_etaria)

contagem_faixa_etaria.plot(kind='bar', color='pink')
plt.xlabel('Faixa Etária')
plt.ylabel('Quantidade')
plt.title('Quantidade de passageiros por faixa etária: ')
```

```python
# Quantidade de Passageiros por Classe
contagem_classe = dataframe_copia['Pclass'].value_counts()
print(contagem_classe)

contagem_classe.plot(kind='bar', color='pink')
plt.xlabel('Classe')
plt.ylabel('Quantidade')
plt.title('Quantidade de passageiros por classe: ')
```

```python
#Quantidade de sobreviventes por classe
quantidade_sobreviventes_classe = dataframe_copia.groupby('Pclass')['Survived'].mean()
print(quantidade_sobreviventes_classe)

quantidade_sobreviventes_classe.plot(kind='bar', color='pink')
plt.xlabel('Classe')
plt.ylabel('Quantidade')
plt.title('Quantidade de sobreviventes por classe')
```

```python
#Quantidade de sobreviventes por gênero
quantidade_sobreviventes_classe = dataframe_copia.groupby('Sex')['Survived'].mean()
print(quantidade_sobreviventes_classe)

quantidade_sobreviventes_classe.plot(kind='bar', color='pink')
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.title('Quantidade de sobreviventes por gênero')
```

```python
#Quantidade de sobreviventes por faixa etária
quantidade_sobreviventes_classe = dataframe_copia.groupby('FaixaEtaria')['Survived'].mean()
print(quantidade_sobreviventes_classe)

quantidade_sobreviventes_classe.plot(kind='bar', color='pink')
plt.xlabel('Faixa Etária')
plt.ylabel('Quantidade')
plt.title('Quantidade de sobreviventes por faixa etária')
```


## **Conclusão**

Após análise dos dados, podemos observar que mesmo que houvessem mais homens do que mulheres à bordo do Titanic, mais mulheres sobreviveram, podemos supor que seja por conta dos direitos deste grupo de pessoas, teoricamente mais frágeis e dependentes e a existência da Declaração sobre a Proteção da Mulher e da Criança em Estados de Emergência e de Conflitos Armados, criada pela ONU em 1974. (ref: https://super.abril.com.br/mundo-estranho/por-que-mulheres-e-criancas-tem-prioridade-em-situacoes-de-emergencia).

Assim, nossa gerente Amanda teria grandes chances de sobreviver se estivesse a bordo do Titanic.
