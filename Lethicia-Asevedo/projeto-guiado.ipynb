## **Criação do Dataframe** 


```python
dataframe = pd.read_csv("Steam_2024_bestRevenue_1500.csv")
dataframe
```

```python
#Retorna o tamanho da tabela : número de linhas x número de colunas
dataframe.shape
```

```python
#A função "ISNULL" te mostra os valores nulos da tabela, mas quando adicionamos a função "sum", ele soma a quantia 
# de valores nulos em cada coluna e apresenta isso em números e nos apresenta de uma forma mais confortável para visualização
# e compreensão dos valores.
dataframe.isnull().sum()
```

```python
#Retorna as 5 primeiras linhas e as 5 últimas linhas da coluna de avaliações
dataframe.reviewScore
```

```python
#Retorna os 10 jogos mais bem avaliados da plataforma
top_reviewScore = dataframe.reviewScore
top_reviewScore = dataframe.nlargest(10, 'reviewScore')

top_reviewScore
```

```python
#Retorna os 10 jogos mais bem avaliados. 
grafico_reviewScore = dataframe.groupby('name')['reviewScore'].sum()
mais_avaliados = grafico_reviewScore.nlargest(10)

mais_avaliados
```

```python
#Exibe os 10 jogos mais bem avaliados em forma de gráficos

mais_avaliados = grafico_reviewScore.nlargest(10)
ax = mais_avaliados.plot(kind='barh', color='blue')
ax.set_xlabel('Jogos')
ax.set_ylabel('Número de cópias vendidas')
ax.set_title('Top 10 jogos mais bem avaliados')
```

```python
#Retorna os 10 jogos com mais cópias vendidas 
top_copieSold = dataframe.copiesSold
top_copieSold = dataframe.nlargest(10, 'copiesSold')

top_copieSold

```

```python
#Exibiu os 10 jogos com mais vendas de cópias. 
grafico_copieSold = dataframe.groupby('name')['copiesSold'].sum()
mais_vendidos = grafico_copieSold.nlargest(10)

mais_vendidos
```

```python
mais_vendidos = grafico_copieSold.nlargest(10)
ax = mais_vendidos.plot(kind='barh', color='blue')
ax.set_xlabel('Jogos')
ax.set_ylabel('Número de cópias vendidas')
ax.set_title('Top 10 jogos com mais cópias vendidas')
```
